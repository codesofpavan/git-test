```python
company_list = [
    r'/workspaces/git-test/Data Analysis with Python/Udemy/individual_stocks_5yr/AAPL_data.csv',
    r'/workspaces/git-test/Data Analysis with Python/Udemy/individual_stocks_5yr/AMZN_data.csv',
    r'/workspaces/git-test/Data Analysis with Python/Udemy/individual_stocks_5yr/GOOG_data.csv',
    r'/workspaces/git-test/Data Analysis with Python/Udemy/individual_stocks_5yr/MSFT_data.csv'

] 
```

```python
company_list    
```

```python
all_data = pd.DataFrame()
for file in company_list:
    current_df = pd.read_csv(file)
    all_data = current_df._append(all_data,ignore_index = True)
```

```python
all_data['date'] = pd.to_datetime(all_data['date'])
```

```python
tech_list = all_data['Name'].unique()
```

```python
tech_list
```

```python
plt.figure(figsize=(20,12))
for index,company in enumerate(tech_list,1):
    plt.subplot(2,2,index)
    filter1 = all_data['Name'] == company
    df = all_data[filter1]
    plt.plot(df['date'],df['close'])
    plt.title(company)
```
